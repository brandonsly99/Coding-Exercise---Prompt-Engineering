# Exercise 3 – Self-Reflection Prompt for Improving Output (No API Key)

Tool Used: Google Colab (Python)

Goal: Create an initial summary, then use a self-reflection prompt to critique it
against explicit criteria and produce an improved summary (before/after).

## Prompt 1 (Before Summary)

Summarize the source text for a general audience in 120–150 words.
Must be neutral and factual. Single paragraph.

## BEFORE SUMMARY

Several cities expanded electric bus fleets in 2024 to cut air pollution and emissions. Transit agencies said electric buses
may reduce fuel and maintenance costs over time, though they cost more to buy than diesel vehicles. Agencies also pointed
to challenges such as installing charging infrastructure and ensuring the electric grid can handle overnight charging for
many buses. Some cities planned to charge buses during off-peak hours to reduce grid strain. Riders often described the
experience as quieter, while operators highlighted the need for technician training and careful route planning, since
cold weather and steep terrain can reduce driving range.

## Prompt 2 (Self-Reflection Critique + Improve)

You are going to critique and improve a summary.

CRITERIA (must check all):
1) Accuracy: no claims not supported by the source text
2) Clarity: simple language, avoid jargon or explain it
3) Audience: general reader
4) Length: 120–150 words
5) Format: single paragraph, no bullets
6) Coverage: include the main point + at least 2 supporting details

Step A — Critique:
List 4–6 concrete issues in the BEFORE SUMMARY referencing the criteria.

Step B — Revised summary:
Write an improved summary that meets ALL criteria.

## CRITIQUE (Step A)

1) Length is likely under the 120–150 word range, so it may not meet the length requirement.
2) It could be clearer about the specific bottleneck (“grid capacity” and “overnight charging”) by explicitly connecting them.
3) It mentions “emissions” but could clarify that the goal includes reducing greenhouse gas emissions and local pollution.
4) It could improve coverage by explicitly mentioning the higher upfront cost as a tradeoff alongside infrastructure needs.
5) It could be slightly more concise in wording while still meeting the length requirement.
6) It should ensure every point directly maps to the source text and avoid adding any extra assumptions.

## AFTER SUMMARY (Step B)

In 2024, several cities expanded electric bus programs to reduce local air pollution and greenhouse gas emissions. Transit agencies reported that electric buses can lower fuel and maintenance costs over time, but they also noted that the upfront purchase price is higher than diesel models. Agencies identified charging infrastructure and grid capacity as major bottlenecks, especially when many buses need to recharge overnight, and some cities planned charging during off-peak hours to reduce strain on the grid. Riders generally described trips as quieter. Operators also emphasized the need for technician training and careful route planning because cold weather and steep routes can reduce driving range.
